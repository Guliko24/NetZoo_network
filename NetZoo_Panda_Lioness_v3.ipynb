# 3. RUN the PANDA

Before running Lioness, we will first need to generate a Panda object. This will be used later to run Lioness. Note that the argument keep_expression_matrix should be specified as True. As Lioness iteractions need to call Panda function to build networks, which needs expression matrix as input.

```python
panda_obj = Panda('/content/netZooPy/tests/puma/ToyData/ToyExpressionData.txt',
                  '/content/netZooPy/tests/puma/ToyData/ToyMotifData.txt',
                  '/content/netZooPy/tests/puma/ToyData/ToyPPIData.txt',
                  remove_missing=False,
                  keep_expression_matrix=True, save_memory=False, modeProcess='legacy')
```

# 4. RUN LIONESS to estimate sample-specific networks


We will first use the Panda object as input for Lioness object. Then Lioness will run Panda algorithm in its iterations to estimate sample-specific network for each sample.




```python
lioness_obj = Lioness(panda_obj)
```

```python
def top_network_plot(self, index=0, top=10, file="network.pdf"):
    """Plots the top network using LIONESS results.

    Args:
        index: The index of the dataframe to plot.
        top: The number of top regulators.
        file: The name of the file where the plot is going to be saved.
    """
    # we added 2 to the index to account for TF and Gene columns
    # Modification: Ensuring the assigned value has a single column
    try:
        # If self.lioness_results.iloc[:, index + 2] is a Series (single column),
        # it can be assigned directly.
        self.export_panda_results['force'] = self.lioness_results.iloc[:, index + 2]
    except ValueError:
        # If it's a DataFrame (multiple columns), select the first column using iloc[:, 0]
        self.export_panda_results['force'] = self.lioness_results.iloc[:, index + 2].iloc[:, 0]

    plot = AnalyzePanda(self)
    plot.top_network_plot(top, file)
```
